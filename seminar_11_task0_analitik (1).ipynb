```python
from sympy import *
from sympy.plotting import plot
import math
init_printing()
```

x = Symbol('x')
f = ((-12 * (x ** 4)) * (sin(cos(x))) - (18 * (x ** 3)) + (5 * (x ** 2)) + 10 * x - 30)
f

### 1. Определить корни

```python
solveset (f, x)
```

### 4. Построить график

```python
plot(f)
print(f)
```

Данная функция является функцией многочленом смешанного типа четвертой степени, она имеет один и тот же предел при стремлении к плюс и к минус бесконечности. Но так как a<0 (-12), то функция убывает до минус бесконечности с обеих сторон, а значит, имеет глобальный максимум.

### 4а. Построить график (ограниченный диапазон)

```python
plot(f, (x, -1.5, 2.5), legend=True)
print(f)
```
